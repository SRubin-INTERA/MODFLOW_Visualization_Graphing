```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
from datetime import timedelta as td
import numpy as np
from pandas import Timestamp

#define filepath for dataset; load excel file to DataFrame "df"
file_path = r'C:/Users/srubin/OneDrive - INTERA Inc/Lisbon/MODFLOW_Visualization/l210610d/l210610d_targets.xlsx'
df = pd.read_excel(file_path)

#define a reference date: 04/01/1990; convert to a "Date" column
refDate = dt(1990, 4, 1)
df['Date'] = refDate + pd.TimedeltaIndex(df['Time'], unit='D')
```

```python
#define series for graphing: observed, computed, layers, residual
observed = df['Observed']
computed = df['Computed']
layers = df['Layer']
residual = df['Residual']
```

```python
#function to create,call,save scatterplot: "Observed_vs_Computed"
def obs_comp_scatter():
    fig, ax = plt.subplots()

    scatter = ax.scatter(observed, computed, c=layers, marker='.', cmap='tab20c', alpha=0.95)
    ax.legend(*scatter.legend_elements(), title="Layers", bbox_to_anchor=(1.0, 1.05))
    plt.title("Observed vs. Computed Target Value")
    plt.xlabel("Observed Value")
    plt.ylabel("Model Value")
    plt.grid()
    plt.savefig("Observed_vs_Computed.png")
    plt.show

obs_comp_scatter()
```

```python
#function to create,call,save scatterplot: "Observed_vs_Residuals"
def obs_resid_scatter():
    fig, ax = plt.subplots()

    scatter = ax.scatter(observed, residual, c=layers, marker='.', cmap='tab20c', alpha=0.95)
    ax.legend(*scatter.legend_elements(), title="Layers", bbox_to_anchor=(1.0, 1.05))
    plt.title("Observed vs. Residuals")
    plt.xlabel("Observed Value")
    plt.ylabel("Residual")
    plt.grid()
    plt.savefig("Observed_vs_Residuals.png")
    plt.show()
    
obs_resid_scatter()
```
